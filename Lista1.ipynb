Dependencies

```python
import sys
sys.path.append(f'../FinanceHub')

from calendars import DayCounts
import pandas as pd
```

Creating the class responsible for calculating business days

```python
# Using BUS/252 and 'anbima' as standards for Brazil
dc = DayCounts('BUS/252', calendar='anbima')
# Reference date February 5th
t = pd.to_datetime('2021-02-05')
expiry_dates = [ '2021-04-01','2021-07-01','2021-10-01','2022-01-01',
                 '2022-04-01','2022-07-01','2022-10-01','2023-01-01',
                 '2023-07-01','2024-01-01','2024-07-01']
du = []
for date in expiry_dates:
    T = pd.to_datetime(date)
    du.append([date, dc.days(t, T)])
output = pd.DataFrame(du, columns=["Expiry Date", "Business Days"])
# Print DU table
output
```

## Exercise 2

```python
def calc_pu(_y, _du):
    return 100000.0/(1+_y)**(_du/252)

# List of yields for each expiry date
yield_rates = [
     0.0205800,0.0238850,0.0299040,0.0344630,0.0401480,
     0.0448470,0.0491370,0.0525000,0.0575190,0.0611500,0.0642470
]

pu = []
for y, d in zip(yield_rates, du):
    pu.append(calc_pu(y, d[1]))
pu_df = pd.DataFrame(pu, columns=["PU"])
output.join(pu_df)
```

### Exercise 3

```python
def calc_yield(_pu, expiry, reference=pd.to_datetime('2021-02-05')):
    _dc = DayCounts('BUS/252', calendar='anbima')
    _du = _dc.days(reference, expiry)
    return (100000.0/_pu)**(252/_du) - 1

# Prices for each expiry date
prices = [
    99701.35, 99076.98, 98100.67, 96994.00, 95586.13,
    94072.42, 92386.89, 90750.40, 87493.99, 84262.83, 81036.02,
]

calc_yields = []
for pu, date in zip(prices, expiry_dates):
    calc_yields.append(calc_yield(pu, date))
calc_yields_df = pd.DataFrame(calc_yields, columns=["Yields"])
output.join(calc_yields_df)
```
